```python
def extract_logs(text):
    fwd_microstep, compute_norm, fwd, optimizer_allgather = [], [], [], []   
    stage_data = []
    for line in text:
        if "[Rank" in line:
            parts = line.split('|')[1:]
            metrics = {p.split(':')[0].strip(): float(p.split(':')[1].strip()) for p in parts if ':' in p}
            
            if 'fwd_microstep' in metrics:  
                # Add derived metrics
                metrics["real_compute_ms"] = (metrics["fwd_microstep"] + metrics["bwd_microstep"])
                metrics["real_comm_ms"] = metrics["bwd_allreduce_microstep"]
                metrics["real_comm_pct"] = (metrics["real_comm_ms"] / (metrics["real_comm_ms"] + metrics["real_compute_ms"] + 1e-8)) * 100
                metrics["total_microstep_time"] = (metrics["fwd_microstep"] + metrics["bwd_microstep"] + 
                                                   metrics.get("bwd_allreduce_microstep", 0) + 
                                                   metrics.get("step_microstep", 0))
                metrics["compute_efficiency"] = (metrics["real_compute_ms"] / metrics["total_microstep_time"]) * 100
                fwd_microstep.append(metrics) 
            elif 'compute_norm' in metrics:
                compute_norm.append(metrics)
            elif 'fwd' in metrics:
                fwd.append(metrics)
            elif ('optimizer_allgather' in metrics) or ('optimizer_step' in metrics):
                optimizer_allgather.append(metrics)
        elif "[Stage" in line:
            # Extract key-value pairs like Loss=10.9, Iter=101.11
            data_pairs = re.findall(r"(\w+)=([\d.]+)", line)
            
            # Extract step number
            step_match = re.search(r"Step\s+(\d+)", line)
            step_number = int(step_match.group(1)) if step_match else None
            
            if data_pairs:
                metrics_dict = {k: float(v) for k, v in data_pairs}
                if step_number is not None:
                    metrics_dict['Step'] = step_number
                
                # Also extract TPS, VRAM, Util if present (they might be in different format)
                tps_match = re.search(r'TPS=([\d.]+)', line)
                vram_match = re.search(r'VRAM=([\d.]+)GB', line)
                util_match = re.search(r'Util=([\d.]+)%', line)
                
                if tps_match:
                    metrics_dict['TPS'] = float(tps_match.group(1))
                if vram_match:
                    metrics_dict['VRAM'] = float(vram_match.group(1))
                if util_match:
                    metrics_dict['Util'] = float(util_match.group(1))
                
                stage_data.append(metrics_dict)
    
    df_micro = pd.DataFrame(fwd_microstep)
    df_stage = pd.DataFrame(stage_data)
    df_fwd = pd.DataFrame(fwd)
    
    return df_micro, df_stage, df_fwd
    
csv_files = {
    0: "stage_0.csv",
    1: "stage_1.csv",
    2: "stage_2.csv",
    3: "stage_3.csv"
}

df_all = []

for stage_id, file in csv_files.items():
    df = pd.read_csv(file, index_col=0)
    df["stage"] = stage_id
    df_all.append(df)

df_all = pd.concat(df_all, ignore_index=True)

logs = ['stage_0_0-480.log',
        'stage_1_0-480.log',
        'stage_2_0-480.log',
        'stage_3_0-480.log']


micro_all = []
stage_all = []
fwd_all = []

for stage_id in range(0, 4):
    with open(logs[stage_id]) as f:
        data = f.readlines()
        df_fwd_microstep, df_stage, df_fwd = extract_logs(data)
        
        # KEY FIX: Assign stage labels here
        df_fwd_microstep['stage'] = stage_id
        df_stage['stage'] = stage_id
        df_fwd['stage'] = stage_id
        
        # KEY FIX: Calculate steps PER STAGE, not globally
        df_fwd_microstep['step'] = df_fwd_microstep.index // 4
        df_fwd['step'] = df_fwd.index * 2
        # (df_stage usually has 'Step' from your regex already)

        micro_all.append(df_fwd_microstep)
        stage_all.append(df_stage)
        fwd_all.append(df_fwd)
        
df_micro_all = pd.concat(micro_all, ignore_index=True) # for each step [0-479]
df_stage_all = pd.concat(stage_all, ignore_index=True) # 1 for 4 steps [0-476]
df_stage_all.columns = ['loss','iteration_time_ms', 'fwd_ms', 'bwd_ms','step_ms', 'throughput_tps', 'peak_gb','gpu_util' , 'step','stage']
df_fwd_all = pd.concat(fwd_all, ignore_index=True) # 1 for 8 steps [0-472]
df_micro_all.shape,df_stage_all.shape,df_fwd_all.shape
```

```python
def process_stage(df, df_rem, prev_df=None):
    # ---- Add missing rows ----
    missing = df_rem[~df_rem['step'].isin(df['step'])]
    missing = missing.reindex(columns=df.columns)
    
    df_updated = pd.concat([df, missing], ignore_index=True)
    df_updated = df_updated.sort_values('step').reset_index(drop=True)

    # ---- Fill with mean ----
    mean_cols = ['alloc_gb', 'reserved_gb']
    for col in mean_cols:
        if col in df_updated.columns:
            df_updated[col] = df_updated[col].fillna(df_updated[col].mean())

    # ---- Interpolate ----
    interp_cols = ['temp_c', 'compute_ms']
    for col in interp_cols:
        if col in df_updated.columns:
            df_updated[col] = df_updated[col].interpolate(method='linear')

    # ---- Fill lr from previous stage ----
    if prev_df is not None and 'lr' in df_updated.columns:
        df_updated['lr'] = df_updated['lr'].fillna(
            df_updated['step'].map(prev_df.set_index('step')['lr'])
        )

    return df_updated
```

```python
df2_updated = process_stage(df2, df2_rem, prev_df=df1)
df3_updated = process_stage(df3, df3_rem, prev_df=df2_updated).bfill()
```

```python
import pandas as pd
import numpy as np

# Load JSON files for all stages
json_files = {
    0: "../results/stage_0_metrics.json",
    1: "../results/stage_1_metrics.json",
    2: "../results/stage_2_metrics.json",
    3: "../results/stage_3_metrics.json",
}

# Load all JSON data into a single DataFrame
df_json_all = []
for stage_id, file in json_files.items():
    df = pd.read_json(file)
    df['stage'] = stage_id
    df_json_all.append(df)

df_json_all = pd.concat(df_json_all, ignore_index=True)

print(f"JSON data shape: {df_json_all.shape}")
print(f"JSON columns: {df_json_all.columns.tolist()}")
print(f"\ndf_stage_all shape: {df_stage_all.shape}")
print(f"df_stage_all columns: {df_stage_all.columns.tolist()}")

# ============================================================================
# COMBINE df_stage_all with JSON data
# ============================================================================

# Start with df_stage_all (keep all its rows)
df_combined = df_stage_all.copy()

# Identify which columns are missing from df_stage_all but present in JSON
json_columns = set(df_json_all.columns) - set(df_stage_all.columns)
print(f"\nColumns in JSON but missing in df_stage_all: {json_columns}")

# Add missing columns as NaN
for col in json_columns:
    df_combined[col] = np.nan

# Now merge/update with JSON data for each stage
for stage_id in [0, 1, 2, 3]:
    # Get JSON data for this stage
    json_stage = df_json_all[df_json_all['stage'] == stage_id].copy()
    
    # Get stage data from combined DataFrame
    stage_mask = df_combined['stage'] == stage_id
    
    # For each column in JSON, update the values in df_combined
    for col in json_stage.columns:
        if col in df_combined.columns and col != 'stage':
            # Create a mapping from step to value
            step_to_value = dict(zip(json_stage['step'], json_stage[col]))
            
            # Update values where step matches
            for step, value in step_to_value.items():
                mask = stage_mask & (df_combined['step'] == step)
                df_combined.loc[mask, col] = value

print(f"\nCombined shape: {df_combined.shape}")
print(f"Combined columns: {df_combined.columns.tolist()}")

# ============================================================================
# INTELLIGENT FILLING OF MISSING VALUES
# ============================================================================

print("\n" + "="*60)
print("INTELLIGENT FILLING OF MISSING VALUES")
print("="*60)

# Check missing values before filling
print("\nMissing values before filling:")
missing_before = df_combined.isnull().sum()
print(missing_before[missing_before > 0])

# Fill based on column type and stage
for stage_id in [0, 1, 2, 3]:
    stage_mask = df_combined['stage'] == stage_id
    
    for col in df_combined.columns:
        if col == 'stage' or col == 'step':
            continue
            
        # Get the column data for this stage
        col_data = df_combined.loc[stage_mask, col]
        
        if col_data.isnull().any():
            print(f"\nFilling {col} for Stage {stage_id}: {col_data.isnull().sum()} missing values")
            
            # Strategy 1: Forward fill within same stage (for temporal data)
            if col in ['loss', 'throughput_tps', 'iteration_time_ms', 'fwd_ms', 'bwd_ms', 
                       'step_ms', 'compute_ms', 'comm_ms', 'comm_pct', 'gpu_util', 
                       'temp_c', 'alloc_gb', 'reserved_gb', 'peak_gb']:
                # Use forward fill within each stage (by step order)
                df_combined.loc[stage_mask, col] = df_combined.loc[stage_mask, col].fillna(method='ffill')
                # Backward fill any remaining at the beginning
                df_combined.loc[stage_mask, col] = df_combined.loc[stage_mask, col].fillna(method='bfill')
                print(f"  → Used forward/backward fill")
            
            # Strategy 2: Fill with stage mean (for metrics that should be stable)
            elif col in ['lr']:
                stage_mean = df_combined.loc[stage_mask, col].mean()
                df_combined.loc[stage_mask, col] = df_combined.loc[stage_mask, col].fillna(stage_mean)
                print(f"  → Filled with stage mean: {stage_mean:.6f}")
            
            # Strategy 3: Fill with 0 (for timing that might be zero)
            elif col in ['comm_ms', 'step_ms']:
                df_combined.loc[stage_mask, col] = df_combined.loc[stage_mask, col].fillna(0)
                print(f"  → Filled with 0")
            
            # Strategy 4: Interpolate for smooth metrics
            elif col in ['loss', 'throughput_tps']:
                # Sort by step first
                df_combined.loc[stage_mask, col] = df_combined.loc[stage_mask].sort_values('step')[col].interpolate(method='linear').values
                print(f"  → Used linear interpolation")
            
            # Strategy 5: Default - fill with stage median
            else:
                stage_median = df_combined.loc[stage_mask, col].median()
                df_combined.loc[stage_mask, col] = df_combined.loc[stage_mask, col].fillna(stage_median)
                print(f"  → Filled with stage median: {stage_median}")

# Check missing values after filling
print("\n" + "="*60)
print("Missing values after filling:")
missing_after = df_combined.isnull().sum()
print(missing_after[missing_after > 0] if any(missing_after > 0) else "✓ No missing values remaining!")

# ============================================================================
# VERIFY THE COMBINED DATA
# ============================================================================

print("\n" + "="*60)
print("VERIFICATION")
print("="*60)

# Check each stage
for stage_id in [0, 1, 2, 3]:
    stage_data = df_combined[df_combined['stage'] == stage_id]
    print(f"\nStage {stage_id}:")
    print(f"  Steps: {stage_data['step'].min()} to {stage_data['step'].max()}")
    print(f"  Rows: {len(stage_data)}")
    print(f"  Columns with data: {stage_data.count().gt(0).sum()} / {len(stage_data.columns)}")
    
    # Sample the first few rows
    print(f"  Sample (first 3 steps):")
    sample_cols = ['step', 'loss', 'throughput_tps', 'fwd_ms', 'bwd_ms', 'comm_pct', 'peak_gb']
    existing_cols = [col for col in sample_cols if col in stage_data.columns]
    print(stage_data[existing_cols].head(3).to_string(index=False))

# ============================================================================
# COMPARE ORIGINAL VS FILLED VALUES
# ============================================================================

print("\n" + "="*60)
print("COMPARISON: ORIGINAL JSON VS COMBINED DATAFRAME")
print("="*60)

for stage_id in [0, 1, 2, 3]:
    json_stage = df_json_all[df_json_all['stage'] == stage_id]
    combined_stage = df_combined[df_combined['stage'] == stage_id]
    
    print(f"\nStage {stage_id}:")
    print(f"  JSON rows: {len(json_stage)}")
    print(f"  Combined rows: {len(combined_stage)}")
    
    # Check if all JSON steps are present in combined
    json_steps = set(json_stage['step'])
    combined_steps = set(combined_stage['step'])
    missing_steps = json_steps - combined_steps
    
    if missing_steps:
        print(f"  ⚠ Steps in JSON but not in combined: {sorted(missing_steps)[:5]}...")
    else:
        print(f"  ✓ All JSON steps present in combined")

# ============================================================================
# SAVE THE COMBINED DATAFRAME
# ============================================================================

df_combined.to_csv('combined_stage_all_with_json.csv', index=False)
print("\n✓ Saved combined DataFrame to 'combined_stage_all_with_json.csv'")

# ============================================================================
# QUICK VERIFICATION PLOT
# ============================================================================

import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Combined Data Verification (df_stage_all + JSON)', fontsize=14, fontweight='bold')

# Plot 1: Loss comparison
ax1 = axes[0, 0]
for stage_id in [0, 1, 2, 3]:
    stage_data = df_combined[df_combined['stage'] == stage_id]
    if 'loss' in stage_data.columns:
        ax1.plot(stage_data['step'], stage_data['loss'], label=f'Stage {stage_id}', linewidth=1.5)
ax1.set_xlabel('Step')
ax1.set_ylabel('Loss')
ax1.set_title('Loss Curves (Combined)')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Plot 2: Throughput comparison
ax2 = axes[0, 1]
for stage_id in [0, 1, 2, 3]:
    stage_data = df_combined[df_combined['stage'] == stage_id]
    if 'throughput_tps' in stage_data.columns:
        ax2.plot(stage_data['step'], stage_data['throughput_tps'], label=f'Stage {stage_id}', linewidth=1.5)
ax2.set_xlabel('Step')
ax2.set_ylabel('Throughput (tokens/sec)')
ax2.set_title('Throughput (Combined)')
ax2.legend()
ax2.grid(True, alpha=0.3)

# Plot 3: Memory usage
ax3 = axes[1, 0]
for stage_id in [0, 1, 2, 3]:
    stage_data = df_combined[df_combined['stage'] == stage_id]
    if 'peak_gb' in stage_data.columns:
        ax3.plot(stage_data['step'], stage_data['peak_gb'], label=f'Stage {stage_id}', linewidth=1.5)
ax3.set_xlabel('Step')
ax3.set_ylabel('Peak Memory (GB)')
ax3.set_title('Memory Usage (Combined)')
ax3.legend()
ax3.grid(True, alpha=0.3)

# Plot 4: Communication overhead
ax4 = axes[1, 1]
for stage_id in [0, 1, 2, 3]:
    stage_data = df_combined[df_combined['stage'] == stage_id]
    if 'comm_pct' in stage_data.columns:
        ax4.plot(stage_data['step'], stage_data['comm_pct'], label=f'Stage {stage_id}', linewidth=1.5)
ax4.set_xlabel('Step')
ax4.set_ylabel('Communication %')
ax4.set_title('Communication Overhead (Combined)')
ax4.legend()
ax4.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('combined_data_verification.png', dpi=150, bbox_inches='tight')
plt.show()

print("\n✓ Verification plots saved to 'combined_data_verification.png'")

# ============================================================================
# FINAL SUMMARY
# ============================================================================

print("\n" + "="*60)
print("FINAL SUMMARY")
print("="*60)
print(f"Total rows in combined DataFrame: {len(df_combined)}")
print(f"Total columns: {len(df_combined.columns)}")
print(f"Stages included: {df_combined['stage'].unique().tolist()}")
print(f"Step range: {df_combined['step'].min()} to {df_combined['step'].max()}")
print(f"\nColumns in final DataFrame:")
for col in df_combined.columns:
    non_null = df_combined[col].notnull().sum()
    print(f"  {col:20s}: {non_null:5d} / {len(df_combined):5d} non-null ({non_null/len(df_combined)*100:.1f}%)")
```

```python
import pandas as pd
import numpy as np

# Your existing data loading
csv_files = {
    0: "stage_0.csv",
    1: "stage_1.csv",
    2: "stage_2.csv",
    3: "stage_3.csv"
}

df_all = []

for stage_id, file in csv_files.items():
    df = pd.read_csv(file, index_col=0)
    df["stage"] = stage_id
    df_all.append(df)

df_all = pd.concat(df_all, ignore_index=True)

logs = ['stage_0_0-480.log',
        'stage_1_0-480.log',
        'stage_2_0-480.log',
        'stage_3_0-480.log']

micro_all = []
stage_all = []
fwd_all = []
for stage_id in range(0, 4):
    with open(logs[stage_id]) as f:
        data = f.readlines()
        df_fwd_microstep, df_stage, df_fwd = extract_logs(data)
        df_fwd_microstep["stage"] = stage_id
        df_stage["stage"] = stage_id
        df_fwd["stage"] = stage_id
        micro_all.append(df_fwd_microstep)
        stage_all.append(df_stage)
        fwd_all.append(df_fwd)

df_micro_all = pd.concat(micro_all, ignore_index=True)
df_stage_all = pd.concat(stage_all, ignore_index=True)
df_fwd_all = pd.concat(fwd_all, ignore_index=True)

print(f"Shapes: {df_micro_all.shape}, {df_stage_all.shape}, {df_fwd_all.shape}")



# ============================================================================
# CREATE AGGREGATED VIEWS
# ============================================================================

# 1. Per-step aggregated (sum of microsteps within each step)
df_step_aggregated = df_micro_all.groupby(['stage', 'step']).agg({
    'fwd_microstep': 'sum',
    'bwd_microstep': 'sum',
    'bwd_inner_microstep': 'sum',
    'bwd_allreduce_microstep': 'sum',
    'step_microstep': 'sum',
    'microstep_idx': 'count'
}).rename(columns={'microstep_idx': 'num_microsteps'}).reset_index()

# Add stage metrics to aggregated view
df_step_aggregated = df_step_aggregated.merge(
    df_stage_all[['step', 'stage', 'Loss', 'TPS', 'VRAM', 'Util', 'FWD', 'BWD', 'STEP']],
    on=['stage', 'step'],
    how='left'
)

# Add fwd data (expanded to match steps)
df_fwd_expanded = df_fwd_all.copy()
# Create entries for all steps by duplicating each fwd record for 2 steps
df_fwd_expanded_list = []
for _, row in df_fwd_expanded.iterrows():
    for offset in [0, 1]:
        new_row = row.copy()
        new_row['step'] = row['step'] + offset
        df_fwd_expanded_list.append(new_row)
df_fwd_expanded = pd.DataFrame(df_fwd_expanded_list)

df_step_aggregated = df_step_aggregated.merge(
    df_fwd_expanded[['step', 'stage', 'fwd', 'bwd', 'bwd_inner', 'bwd_allreduce']],
    on=['stage', 'step'],
    how='left'
)

# Add CSV metrics
df_step_aggregated = df_step_aggregated.merge(
    df_all[['step', 'stage', 'throughput_tps', 'loss', 'iteration_time_ms', 
            'fwd_ms', 'bwd_ms', 'step_ms', 'compute_ms', 'comm_ms', 
            'comm_pct', 'gpu_util', 'peak_gb']],
    on=['stage', 'step'],
    how='left'
)

# ============================================================================
# VERIFY THE COMBINED DATA
# ============================================================================

print("\n" + "="*60)
print("DATA VERIFICATION")
print("="*60)

print(f"\ndf_micro_all: {len(df_micro_all)} rows (microsteps)")
print(f"  - Steps per stage: {df_micro_all.groupby('stage')['step'].nunique().to_dict()}")
print(f"  - Microsteps per step: {df_micro_all.groupby(['stage', 'step']).size().iloc[0]}")

print(f"\ndf_stage_all: {len(df_stage_all)} rows (iterations)")
print(f"  - Steps range: {df_stage_all['step'].min()} to {df_stage_all['step'].max()}")

print(f"\ndf_fwd_all: {len(df_fwd_all)} rows (optimizer steps)")
print(f"  - Steps range: {df_fwd_all['step'].min()} to {df_fwd_all['step'].max()}")

print(f"\ndf_final: {len(df_final)} rows")
print(f"  - Columns: {df_final.columns.tolist()}")
print(f"  - Null values per column:\n{df_final.isnull().sum()}")

print(f"\ndf_step_aggregated: {len(df_step_aggregated)} rows")
print(f"  - Columns: {df_step_aggregated.columns.tolist()}")

# ============================================================================
# SAVE COMBINED DATAFRAMES
# ============================================================================

df_final.to_csv('combined_all_data.csv', index=False)
df_step_aggregated.to_csv('step_aggregated_data.csv', index=False)
df_micro_all.to_csv('microstep_data.csv', index=False)
df_stage_all.to_csv('iteration_data.csv', index=False)
df_fwd_all.to_csv('optimizer_step_data.csv', index=False)

print("\n✓ Saved all combined DataFrames to CSV files")

# ============================================================================
# QUICK VERIFICATION PLOTS
# ============================================================================

import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# 1. Verify microstep alignment
ax1 = axes[0, 0]
stage0_micro = df_micro_all[df_micro_all['stage'] == 0]
stage0_micro_Step0 = stage0_micro[stage0_micro['step'] == 0]
ax1.bar(range(4), stage0_micro_Step0['fwd_microstep'], label='FWD')
ax1.bar(range(4), stage0_micro_Step0['bwd_microstep'], bottom=stage0_micro_Step0['fwd_microstep'], label='BWD')
ax1.set_xlabel('Microstep within Step 0')
ax1.set_ylabel('Time (ms)')
ax1.set_title('Microstep Breakdown for Step 0 (Stage 0)')
ax1.legend()
ax1.grid(True, alpha=0.3)

# 2. Compare loss from different sources
ax2 = axes[0, 1]
stage0_stage = df_stage_all[df_stage_all['stage'] == 0]
stage0_csv = df_all[df_all['stage'] == 0]
ax2.plot(stage0_stage['step'], stage0_stage['Loss'], 'o-', label='From Logs', markersize=3)
ax2.plot(stage0_csv['step'], stage0_csv['loss'], 's--', label='From CSV', markersize=3)
ax2.set_xlabel('Step')
ax2.set_ylabel('Loss')
ax2.set_title('Loss Comparison: Logs vs CSV')
ax2.legend()
ax2.grid(True, alpha=0.3)

# 3. Compare TPS
ax3 = axes[1, 0]
ax3.plot(stage0_stage['step'], stage0_stage['TPS'], 'o-', label='From Logs', markersize=3)
ax3.plot(stage0_csv['step'], stage0_csv['throughput_tps'], 's--', label='From CSV', markersize=3)
ax3.set_xlabel('Step')
ax3.set_ylabel('Throughput (tokens/sec)')
ax3.set_title('Throughput Comparison: Logs vs CSV')
ax3.legend()
ax3.grid(True, alpha=0.3)

# 4. Forward time comparison
ax4 = axes[1, 1]
stage0_fwd = df_fwd_all[df_fwd_all['stage'] == 0]
ax4.plot(stage0_fwd['step'], stage0_fwd['fwd'], 'o-', label='From fwd_all', markersize=4)
ax4.plot(stage0_csv['step'], stage0_csv['fwd_ms'], 's--', label='From CSV', markersize=4)
ax4.set_xlabel('Step')
ax4.set_ylabel('Forward Time (ms)')
ax4.set_title('Forward Time Comparison')
ax4.legend()
ax4.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('combined_data_verification.png', dpi=150, bbox_inches='tight')
plt.show()

print("\n✓ Verification plots saved to 'combined_data_verification.png'")

# ============================================================================
# PRINT SAMPLE OF COMBINED DATA
# ============================================================================

print("\n" + "="*60)
print("SAMPLE OF COMBINED DATA (First 5 rows, Stage 0, Step 0)")
print("="*60)

sample = df_final[(df_final['stage'] == 0) & (df_final['step'] == 0)].head(5)
print(sample[['stage', 'step', 'microstep_idx', 'fwd_microstep', 'bwd_microstep', 
              'Loss', 'TPS', 'fwd', 'throughput_tps', 'loss']].to_string())
```

```python

print(f"Microsteps per step by stage:\n{microsteps_per_step}")

# Add step number to microstep data
# Assuming microsteps are in order: steps 1,2,3... each with N microsteps
def add_step_numbers(df_micro, steps_per_stage):
    """
    Add step numbers to microstep data based on consistent microstep count per step
    """
    df_micro_with_step = []
    
    for stage_id in df_micro['stage'].unique():
        df_stage = df_micro[df_micro['stage'] == stage_id].copy()
        n_steps = len(df_all[df_all['stage'] == stage_id])
        n_microsteps = len(df_stage)
        
        if n_steps > 0:
            micro_per_step = n_microsteps // n_steps
            df_stage['step'] = (df_stage.groupby('stage').cumcount() // micro_per_step) + 1
            df_micro_with_step.append(df_stage)
    
    return pd.concat(df_micro_with_step, ignore_index=True)

df_micro_all = add_step_numbers(df_micro_all, None)
```

```python


# Verify alignment
print("\nStep alignment verification:")
for stage in df_micro_all['stage'].unique():
    micro_steps = df_micro_all[df_micro_all['stage'] == stage]['step'].nunique()
    df_steps = df_all[df_all['stage'] == stage]['step'].nunique()
    print(f"Stage {stage}: Micro steps: {micro_steps}, DF steps: {df_steps}")

# ============================================================================
# AGGREGATE MICROSTEP DATA TO STEP LEVEL
# ============================================================================

# Aggregate microstep metrics per step
micro_step_agg = df_micro_all.groupby(['stage', 'step']).agg({
    'fwd_microstep': 'mean',
    'bwd_microstep': 'mean',
    'bwd_inner_microstep': 'mean',
    'bwd_allreduce_microstep': 'mean',
    'step_microstep': 'mean',
    'real_compute_ms': 'mean',
    'real_comm_ms': 'mean',
    'real_comm_pct': 'mean',
    'total_microstep_time': 'mean',
    'compute_efficiency': 'mean'
}).reset_index()

# Rename columns to avoid confusion
micro_step_agg = micro_step_agg.rename(columns={
    'fwd_microstep': 'micro_fwd_ms',
    'bwd_microstep': 'micro_bwd_ms',
    'bwd_inner_microstep': 'micro_bwd_inner_ms',
    'bwd_allreduce_microstep': 'micro_allreduce_ms',
    'step_microstep': 'micro_step_ms',
    'real_compute_ms': 'micro_compute_ms',
    'real_comm_ms': 'micro_comm_ms',
    'real_comm_pct': 'micro_comm_pct',
    'total_microstep_time': 'micro_total_time_ms',
    'compute_efficiency': 'micro_compute_efficiency_pct'
})

# ============================================================================
# MERGE WITH DF_ALL
# ============================================================================

# Ensure step and stage columns are same type
df_all['step'] = df_all['step'].astype(int)
micro_step_agg['step'] = micro_step_agg['step'].astype(int)

# Merge the aggregated microstep data
df_all = df_all.merge(micro_step_agg, on=['stage', 'step'], how='left')

print("\n" + "="*80)
print("Columns added to df_all from microstep data:")
print("="*80)
added_cols = [col for col in df_all.columns if col.startswith('micro_')]
for col in added_cols:
    print(f"  - {col}")

# ============================================================================
# CALCULATE DERIVED COMMUNICATION METRICS
# ============================================================================

# Communication ratios
df_all["comm_to_compute_ratio"] = df_all["comm_ms"] / (df_all["compute_ms"] + 1e-8)
df_all["micro_comm_to_compute_ratio"] = df_all["micro_comm_ms"] / (df_all["micro_compute_ms"] + 1e-8)

# Difference between main and microstep measurements
df_all["comm_pct_diff"] = df_all["comm_pct"] - df_all["micro_comm_pct"]
df_all["comm_ms_diff"] = df_all["comm_ms"] - df_all["micro_comm_ms"]
df_all["compute_ms_diff"] = df_all["compute_ms"] - df_all["micro_compute_ms"]

# AllReduce contribution
df_all["allreduce_pct_of_step"] = (df_all["micro_allreduce_ms"] / df_all["iteration_time_ms"]) * 100
df_all["allreduce_pct_of_comm"] = (df_all["micro_allreduce_ms"] / (df_all["comm_ms"] + 1e-8)) * 100

# Step time breakdown from microstep data
df_all["micro_fwd_pct"] = (df_all["micro_fwd_ms"] / df_all["micro_total_time_ms"]) * 100
df_all["micro_bwd_pct"] = (df_all["micro_bwd_ms"] / df_all["micro_total_time_ms"]) * 100
df_all["micro_allreduce_pct"] = (df_all["micro_allreduce_ms"] / df_all["micro_total_time_ms"]) * 100

# Overlap estimation (difference between sum of components and total time)
df_all["component_sum_ms"] = (df_all["micro_fwd_ms"] + df_all["micro_bwd_ms"] + 
                               df_all["micro_allreduce_ms"] + df_all["micro_step_ms"])
df_all["overlap_overhead_ms"] = df_all["iteration_time_ms"] - df_all["component_sum_ms"]
df_all["overlap_overhead_pct"] = (df_all["overlap_overhead_ms"] / df_all["iteration_time_ms"]) * 100

# Communication efficiency
df_all["comm_efficiency_pct"] = 100 - df_all["comm_pct"]
df_all["micro_comm_efficiency_pct"] = 100 - df_all["micro_comm_pct"]

# ============================================================================
# VERIFY THE MERGED DATA
# ============================================================================

print("\n" + "="*80)
print("VERIFICATION: Sample of merged data")
print("="*80)

display_cols = [
    'stage', 'step', 'iteration_time_ms', 'comm_pct', 'micro_comm_pct',
    'comm_pct_diff', 'micro_allreduce_ms', 'allreduce_pct_of_step',
    'micro_compute_efficiency_pct'
]

# Show first few rows
print(df_all[display_cols].head(10).round(2))

# Check for missing values
print("\nMissing values in microstep columns:")
missing_counts = df_all[added_cols].isnull().sum()
print(missing_counts[missing_counts > 0])

# ============================================================================
# CREATE SUMMARY TABLE WITH MICROSTEP METRICS
# ============================================================================

print("\n" + "="*80)
print("SUMMARY TABLE: Communication Metrics by Stage")
print("="*80)

summary = df_all.groupby('stage').agg({
    'iteration_time_ms': 'mean',
    'comm_pct': 'mean',
    'micro_comm_pct': 'mean',
    'comm_pct_diff': 'mean',
    'micro_allreduce_ms': 'mean',
    'allreduce_pct_of_step': 'mean',
    'allreduce_pct_of_comm': 'mean',
    'micro_compute_efficiency_pct': 'mean',
    'overlap_overhead_pct': 'mean',
    'micro_fwd_pct': 'mean',
    'micro_bwd_pct': 'mean',
    'micro_allreduce_pct': 'mean'
}).round(2)

summary.columns = [
    'Iter Time (ms)', 'Comm % (Main)', 'Comm % (Micro)', 'Comm Diff (pp)',
    'AllReduce (ms)', 'AllReduce % of Step', 'AllReduce % of Comm',
    'Compute Eff %', 'Overlap Overhead %',
    'FWD %', 'BWD %', 'AllReduce %'
]

print(summary)

# ============================================================================
# EXPORT ENHANCED DATAFRAME
# ============================================================================

df_all.to_csv("df_all_with_microstep_metrics.csv", index=False)
print("\n" + "="*80)
print("Enhanced DataFrame saved to: df_all_with_microstep_metrics.csv")
print("="*80)

# ============================================================================
# PLOT: COMMUNICATION METRICS COMPARISON
# ============================================================================

import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Communication Metrics Across ZeRO Stages (Step-Level Alignment)', fontsize=14, fontweight='bold')

stages = sorted(df_all['stage'].unique())
x = np.arange(len(stages))

# Plot 1: Main vs Microstep Communication %
ax1 = axes[0, 0]
main_comm = [df_all[df_all['stage'] == s]['comm_pct'].mean() for s in stages]
micro_comm = [df_all[df_all['stage'] == s]['micro_comm_pct'].mean() for s in stages]

width = 0.35
ax1.bar(x - width/2, main_comm, width, label='Main (every 4 steps)', color='steelblue')
ax1.bar(x + width/2, micro_comm, width, label='Microstep (every step)', color='coral')
ax1.set_xlabel('ZeRO Stage')
ax1.set_ylabel('Communication Percentage (%)')
ax1.set_title('Comm %: Main vs Microstep')
ax1.set_xticks(x)
ax1.set_xticklabels(stages)
ax1.legend()
ax1.grid(True, alpha=0.3)

# Plot 2: AllReduce Contribution
ax2 = axes[0, 1]
allreduce_pct = [df_all[df_all['stage'] == s]['allreduce_pct_of_step'].mean() for s in stages]
ax2.bar(x, allreduce_pct, color='mediumseagreen', edgecolor='black')
ax2.set_xlabel('ZeRO Stage')
ax2.set_ylabel('AllReduce % of Step Time')
ax2.set_title('AllReduce Contribution to Step Time')
ax2.set_xticks(x)
ax2.set_xticklabels(stages)
ax2.grid(True, alpha=0.3)

# Plot 3: Compute Efficiency
ax3 = axes[1, 0]
compute_eff = [df_all[df_all['stage'] == s]['micro_compute_efficiency_pct'].mean() for s in stages]
ax3.plot(x, compute_eff, 'o-', linewidth=2, markersize=8, color='green')
ax3.set_xlabel('ZeRO Stage')
ax3.set_ylabel('Compute Efficiency (%)')
ax3.set_title('Compute Efficiency (Microstep)')
ax3.set_xticks(x)
ax3.set_xticklabels(stages)
ax3.grid(True, alpha=0.3)
ax3.set_ylim(0, 100)

# Plot 4: Overlap Overhead
ax4 = axes[1, 1]
overlap = [df_all[df_all['stage'] == s]['overlap_overhead_pct'].mean() for s in stages]
ax4.bar(x, overlap, color='purple', edgecolor='black')
ax4.set_xlabel('ZeRO Stage')
ax4.set_ylabel('Overlap Overhead (%)')
ax4.set_title('Overlap Overhead (Component Sum vs Total Time)')
ax4.set_xticks(x)
ax4.set_xticklabels(stages)
ax4.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('communication_metrics_step_aligned.png', dpi=150, bbox_inches='tight')
plt.show()

print("\nPlot saved: communication_metrics_step_aligned.png")

# ============================================================================
# STATISTICAL SUMMARY
# ============================================================================

print("\n" + "="*80)
print("STATISTICAL SUMMARY")
print("="*80)

print("\nCorrelation between Main and Microstep Communication %:")
for stage in stages:
    stage_data = df_all[df_all['stage'] == stage]
    if len(stage_data) > 1:
        corr = stage_data['comm_pct'].corr(stage_data['micro_comm_pct'])
        print(f"Stage {stage}: r = {corr:.3f}")

print("\nAverage difference between Main and Microstep Comm %:")
for stage in stages:
    diff = df_all[df_all['stage'] == stage]['comm_pct_diff'].mean()
    print(f"Stage {stage}: {diff:+.2f} percentage points")
```

```python
def extract_logs(text):
    fwd_microstep, compute_norm, fwd, optimizer_allgather = [], [], [], []   
    for line in text:
        if "[Rank" in line:
            parts = line.split('|')[1:]
            metrics = {p.split(':')[0].strip(): float(p.split(':')[1].strip()) for p in parts if ':' in p}
            
            if 'fwd_microstep' in metrics:  
                # Add derived metrics
                metrics["real_compute_ms"] = (metrics["fwd_microstep"] + metrics["bwd_microstep"])
                metrics["real_comm_ms"] = metrics["bwd_allreduce_microstep"]
                metrics["real_comm_pct"] = (metrics["real_comm_ms"] / (metrics["real_comm_ms"] + metrics["real_compute_ms"] + 1e-8)) * 100
                metrics["total_microstep_time"] = (metrics["fwd_microstep"] + metrics["bwd_microstep"] + 
                                                   metrics.get("bwd_allreduce_microstep", 0) + 
                                                   metrics.get("step_microstep", 0))
                metrics["compute_efficiency"] = (metrics["real_compute_ms"] / metrics["total_microstep_time"]) * 100
                fwd_microstep.append(metrics) 
            elif 'compute_norm' in metrics:
                compute_norm.append(metrics)
            elif 'fwd' in metrics:
                fwd.append(metrics)
            elif ('optimizer_allgather' in metrics) or ('optimizer_step' in metrics):
                optimizer_allgather.append(metrics)
                
    return fwd_microstep, compute_norm, fwd, optimizer_allgather

csv_files = {
    0: "stage_0.csv",
    1: "stage_1.csv",
    2: "stage_2.csv",
    3: "stage_3.csv"
}

df_all = []

for stage_id, file in csv_files.items():
    df = pd.read_csv(file, index_col=0)
    df["stage"] = stage_id
    df_all.append(df)

df_all = pd.concat(df_all, ignore_index=True)

logs = ['stage_0_0-480.log',
        'stage_1_0-480.log',
        'stage_2_0-480.log',
        'stage_3_0-480.log']

micro_all = []
for stage_id in range(0, 4):
    with open(logs[stage_id]) as f:
        data = f.readlines()
        fwd_microstep, compute_norm, fwd, optimizer_allgather = extract_logs(data)
        df_fwd_microstep = pd.DataFrame(fwd_microstep)
        if not df_fwd_microstep.empty:
            df_fwd_microstep["stage"] = stage_id
            if 'microstep_idx' not in df_fwd_microstep.columns:
                df_fwd_microstep['microstep_idx'] = range(len(df_fwd_microstep))
            micro_all.append(df_fwd_microstep)

df_micro_all = pd.concat(micro_all, ignore_index=True)

# Print available columns
print("Microstep DataFrame columns:", df_micro_all.columns.tolist())

# Calculate communication percentage from microstep data only
micro_comm_summary = df_micro_all.groupby("stage").agg({
    "real_comm_pct": "mean"
}).reset_index()
micro_comm_summary.columns = ["stage", "comm_pct_from_microstep"]

print("\nMicrostep Communication Percentage Summary:")
print(micro_comm_summary)


# df_all.columns
# Index(['stage', 'step', 'loss', 'lr', 'iteration_time_ms', 'fwd_ms', 'bwd_ms',
#        'step_ms', 'compute_ms', 'comm_ms', 'comm_pct', 'throughput_tps',
#        'gpu_util', 'temp_c', 'alloc_gb', 'reserved_gb', 'peak_gb',
#        'compute_comm_ratio', 'total_step_time_ms', 'compute_pct'],
#       dtype='object')


# df_micro_all.columns
# Index(['fwd_microstep', 'bwd_microstep', 'bwd_inner_microstep',
#        'bwd_allreduce_microstep', 'step_microstep', 'real_compute_ms',
#        'real_comm_ms', 'real_comm_pct', 'total_microstep_time',
#        'compute_efficiency', 'stage', 'microstep_idx'],
#       dtype='object')
```

```python
import pandas as pd
import numpy as np

# ============================================================================
# MERGE MICROSTEP DATA INTO MAIN DATAFRAME
# ============================================================================

# Since df_all and df_micro_all have different row counts (step-level vs microstep-level),
# we need to aggregate microstep data to step level before merging

# First, add step information to df_micro_all if not present
# The logs don't have step numbers directly, so we need to infer or aggregate

# Option 1: Aggregate microstep data per stage (overall averages)
micro_aggregated = df_micro_all.groupby("stage").agg({
    "fwd_microstep": "mean",
    "bwd_microstep": "mean",
    "bwd_inner_microstep": "mean",
    "bwd_allreduce_microstep": "mean",
    "step_microstep": "mean",
    "real_compute_ms": "mean",
    "real_comm_ms": "mean",
    "real_comm_pct": "mean",
    "total_microstep_time": "mean",
    "compute_efficiency": "mean"
}).reset_index()

# Rename columns to avoid confusion
micro_aggregated = micro_aggregated.rename(columns={
    "fwd_microstep": "micro_fwd_ms",
    "bwd_microstep": "micro_bwd_ms",
    "bwd_inner_microstep": "micro_bwd_inner_ms",
    "bwd_allreduce_microstep": "micro_allreduce_ms",
    "step_microstep": "micro_step_ms",
    "real_compute_ms": "micro_compute_ms",
    "real_comm_ms": "micro_comm_ms",
    "real_comm_pct": "micro_comm_pct",
    "total_microstep_time": "micro_total_time_ms",
    "compute_efficiency": "micro_compute_efficiency_pct"
})

# Merge aggregated microstep data into df_all
df_all = df_all.merge(micro_aggregated, on="stage", how="left")

print("Columns added to df_all from microstep data:")
print([col for col in df_all.columns if col.startswith("micro_")])

# ============================================================================
# CALCULATE ADDITIONAL COMMUNICATION METRICS
# ============================================================================

# Communication ratios
df_all["comm_to_compute_ratio"] = df_all["comm_ms"] / (df_all["compute_ms"] + 1e-8)
df_all["micro_comm_to_compute_ratio"] = df_all["micro_comm_ms"] / (df_all["micro_compute_ms"] + 1e-8)

# Communication efficiency (inverse of overhead)
df_all["comm_efficiency_pct"] = 100 - df_all["comm_pct"]
df_all["micro_comm_efficiency_pct"] = 100 - df_all["micro_comm_pct"]

# Difference between main and microstep measurements
df_all["comm_pct_diff"] = df_all["comm_pct"] - df_all["micro_comm_pct"]
df_all["compute_ms_diff"] = df_all["compute_ms"] - df_all["micro_compute_ms"]
df_all["comm_ms_diff"] = df_all["comm_ms"] - df_all["micro_comm_ms"]

# AllReduce contribution to total time
df_all["allreduce_pct_of_step"] = (df_all["micro_allreduce_ms"] / df_all["iteration_time_ms"]) * 100
df_all["allreduce_pct_of_comm"] = (df_all["micro_allreduce_ms"] / (df_all["comm_ms"] + 1e-8)) * 100

# Step time breakdown from microstep data
df_all["micro_fwd_pct"] = (df_all["micro_fwd_ms"] / df_all["micro_total_time_ms"]) * 100
df_all["micro_bwd_pct"] = (df_all["micro_bwd_ms"] / df_all["micro_total_time_ms"]) * 100
df_all["micro_allreduce_pct"] = (df_all["micro_allreduce_ms"] / df_all["micro_total_time_ms"]) * 100

# Overlap estimation (difference between sum of components and total time)
df_all["component_sum_ms"] = (df_all["micro_fwd_ms"] + df_all["micro_bwd_ms"] + 
                               df_all["micro_allreduce_ms"] + df_all["micro_step_ms"])
df_all["overlap_overhead_ms"] = df_all["iteration_time_ms"] - df_all["component_sum_ms"]
df_all["overlap_overhead_pct"] = (df_all["overlap_overhead_ms"] / df_all["iteration_time_ms"]) * 100

# ============================================================================
# VERIFY ADDED COLUMNS
# ============================================================================

print("\n" + "="*80)
print("UPDATED DF_ALL COLUMNS")
print("="*80)
print(df_all.columns.tolist())

# ============================================================================
# DISPLAY SAMPLE DATA
# ============================================================================

print("\n" + "="*80)
print("SAMPLE DATA WITH NEW COMMUNICATION METRICS")
print("="*80)

display_cols = [
    'stage', 'step', 'iteration_time_ms', 'comm_pct', 'micro_comm_pct',
    'comm_pct_diff', 'micro_allreduce_ms', 'allreduce_pct_of_step',
    'micro_compute_efficiency_pct', 'compute_efficiency_pct'
]

print(df_all[display_cols].head(10).round(2))

# ============================================================================
# AGGREGATE SUMMARY WITH COMMUNICATION METRICS
# ============================================================================

print("\n" + "="*80)
print("COMMUNICATION METRICS SUMMARY BY STAGE")
print("="*80)

comm_summary = df_all.groupby("stage").agg({
    "comm_pct": "mean",
    "micro_comm_pct": "mean",
    "comm_pct_diff": "mean",
    "micro_allreduce_ms": "mean",
    "allreduce_pct_of_step": "mean",
    "allreduce_pct_of_comm": "mean",
    "micro_compute_efficiency_pct": "mean",
    "comm_efficiency_pct": "mean",
    "overlap_overhead_pct": "mean",
    "micro_fwd_pct": "mean",
    "micro_bwd_pct": "mean",
    "micro_allreduce_pct": "mean"
}).round(2)

comm_summary.columns = [
    "Comm % (Main)", "Comm % (Micro)", "Comm Diff (pp)",
    "AllReduce (ms)", "AllReduce % of Step", "AllReduce % of Comm",
    "Compute Eff % (Micro)", "Comm Eff %", "Overlap Overhead %",
    "FWD % (Micro)", "BWD % (Micro)", "AllReduce % (Micro)"
]

print(comm_summary)

# ============================================================================
# EXPORT ENHANCED DATAFRAME
# ============================================================================

df_all.to_csv("df_all_with_communication_metrics.csv", index=False)
print("\n" + "="*80)
print("Enhanced DataFrame saved to: df_all_with_communication_metrics.csv")
print("="*80)

# ============================================================================
# TABLE: COMMUNICATION BREAKDOWN BY STAGE
# ============================================================================

print("\n" + "="*80)
print("TABLE: Communication Breakdown by ZeRO Stage")
print("="*80)

comm_breakdown = df_all.groupby("stage").agg({
    "iteration_time_ms": "mean",
    "compute_ms": "mean",
    "comm_ms": "mean",
    "micro_fwd_ms": "mean",
    "micro_bwd_ms": "mean",
    "micro_allreduce_ms": "mean",
    "micro_step_ms": "mean",
    "micro_comm_pct": "mean",
    "micro_compute_efficiency_pct": "mean"
}).round(2)

comm_breakdown.columns = [
    "Iter Time (ms)", "Compute (ms)", "Comm (ms)",
    "FWD (ms)", "BWD (ms)", "AllReduce (ms)", "Step (ms)",
    "Comm %", "Compute Eff %"
]

print(comm_breakdown)

# ============================================================================
# PLOT: COMMUNICATION METRICS COMPARISON
# ============================================================================

import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Communication Metrics Comparison Across ZeRO Stages', fontsize=14, fontweight='bold')

# Plot 1: Main vs Microstep Communication %
ax1 = axes[0, 0]
stages = df_all.groupby("stage")["stage"].first().values
main_comm = df_all.groupby("stage")["comm_pct"].mean().values
micro_comm = df_all.groupby("stage")["micro_comm_pct"].mean().values

x = np.arange(len(stages))
width = 0.35
ax1.bar(x - width/2, main_comm, width, label='Main', color='steelblue')
ax1.bar(x + width/2, micro_comm, width, label='Microstep', color='coral')
ax1.set_xlabel('ZeRO Stage')
ax1.set_ylabel('Communication Percentage (%)')
ax1.set_title('Comm %: Main vs Microstep')
ax1.set_xticks(x)
ax1.set_xticklabels(stages)
ax1.legend()
ax1.grid(True, alpha=0.3)

# Plot 2: AllReduce Time and Percentage
ax2 = axes[0, 1]
allreduce_ms = df_all.groupby("stage")["micro_allreduce_ms"].mean().values
allreduce_pct = df_all.groupby("stage")["allreduce_pct_of_step"].mean().values

ax2.bar(x, allreduce_ms, color='mediumseagreen', edgecolor='black')
ax2.set_xlabel('ZeRO Stage')
ax2.set_ylabel('AllReduce Time (ms)', color='green')
ax2.tick_params(axis='y', labelcolor='green')
ax2.set_xticks(x)
ax2.set_xticklabels(stages)

ax2_twin = ax2.twinx()
ax2_twin.plot(x, allreduce_pct, 'o-', color='blue', linewidth=2, markersize=8)
ax2_twin.set_ylabel('AllReduce % of Step', color='blue')
ax2_twin.tick_params(axis='y', labelcolor='blue')
ax2.set_title('AllReduce Time and Contribution')
ax2.grid(True, alpha=0.3)

# Plot 3: Compute Efficiency
ax3 = axes[1, 0]
compute_eff = df_all.groupby("stage")["micro_compute_efficiency_pct"].mean().values
comm_eff = df_all.groupby("stage")["comm_efficiency_pct"].mean().values

ax3.plot(x, compute_eff, 'o-', linewidth=2, markersize=8, label='Compute Efficiency (Micro)', color='green')
ax3.plot(x, comm_eff, 's-', linewidth=2, markersize=8, label='Communication Efficiency', color='red')
ax3.set_xlabel('ZeRO Stage')
ax3.set_ylabel('Efficiency (%)')
ax3.set_title('Compute vs Communication Efficiency')
ax3.set_xticks(x)
ax3.set_xticklabels(stages)
ax3.legend()
ax3.grid(True, alpha=0.3)
ax3.set_ylim(0, 100)

# Plot 4: Overlap Overhead
ax4 = axes[1, 1]
overlap = df_all.groupby("stage")["overlap_overhead_pct"].mean().values

ax4.bar(x, overlap, color='purple', edgecolor='black')
ax4.set_xlabel('ZeRO Stage')
ax4.set_ylabel('Overlap Overhead (%)')
ax4.set_title('Overlap Overhead (Component Sum vs Total Time)')
ax4.set_xticks(x)
ax4.set_xticklabels(stages)
ax4.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('communication_metrics_comparison.png', dpi=150, bbox_inches='tight')
plt.show()

print("\nPlot saved: communication_metrics_comparison.png")
```
